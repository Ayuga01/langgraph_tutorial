```python
from langgraph.store.memory import InMemoryStore
```

```python
store = InMemoryStore()
```

```python
#creating namespace
namespace = ("user", "u1")
```

Creating Memory

```python
store.put(namespace, "1", {"data": "User likes pizza"})
store.put(namespace, "2", {"data": "User prefers dark mode"})
```

```python
namespace2 = ("user", "u2")

store.put(namespace2, "1", {"data": "User likes burger"})
store.put(namespace2, "2", {"data": "User prefers grid style naviagtion"})
```

# Retrieving the memory


```python
store.get(namespace2, "1")
```

# Retrieving all memories

```python
items = store.search(namespace)

for item in items:
    print(item.value)
```

# Semantic search

```python
from langchain_huggingface import HuggingFaceEmbeddings
from dotenv import load_dotenv

load_dotenv()

embeddings = HuggingFaceEmbeddings(model='sentence-transformers/all-MiniLM-L6-v2')
```

```python
store = InMemoryStore(index={'embed': embeddings, 'dims':1536})
```

```python
namespace = ("user", "u1")
```

```python
store.put(namespace, "1", {"data": "User prefers long answers"})
store.put(namespace, "2", {"data": "User likes examples in C++"})
store.put(namespace, "3", {"data": "User works late at night"})
store.put(namespace, "4", {"data": "User is learning about vector databases"})
store.put(namespace, "5", {"data": "User enjoys hiking"})
store.put(namespace, "6", {"data": "User is interested in machine learning"})
store.put(namespace, "7", {"data": "User has a cat named Whiskers"})
store.put(namespace, "8", {"data": "User's favorite color is blue"})
store.put(namespace, "9", {"data": "User is a software engineer"})
store.put(namespace, "10", {"data": "User is planning a trip to Japan"})
```

```python
items = store.search(namespace, query="What do for fun?", limit=3)

for item in items:
    print(item.value)
```
